```python
#THIS PROGRAM CREATES AN ACCOUNT FOR THE USERS AND SAVES THE USER DETAILS IN A A FILE WHICH I CAN LATER ACCESS
def details():
    print("\t\t WELCOME TO AKOREDE MICROFINANCE BANK")
    print("\t\t Kindly create an account to proceed")
    Lastname=(input("LASTNAME: ")+"\n")
    Name=(input("FIRSTNAME: ")+"\n")
    Middlename=(input('MIDDLENAME  ')+"\n")
    nickname=(input("What will you like us to call you:")+"\n")
    phone=(input('Input your phone number: ')+"\n")
    account_no=str(phone)
    age=(input("Please input your age  ")+"\n")
    balance="99999"
    filename=input("Kindly create a username to access your account")
    pin=str(input("Please create a 4 digit transaction pin  "))
    while True:
        pin2=(input("Confirm Pin  "))
        if pin==pin2:
            break
        else:
            print("pin doesn't match")
            continue
            clear_output()
    print("Account created successfully")
    print(f"Your account number is {account_no}")
    file1=open(filename,"a")
    file1.write(Lastname)#line0
    file1.write(Name)#line 1
    file1.write(Middlename)#line2
    file1.write(nickname)#line3
    file1.write(phone)#line4070
    file1.write(account_no)#line5
    file1.write(age)#line6
    file1.write(balance)#line7
    file1.write(f"\n{pin}")#line8
    file1.close
details()
```
